```python
import pandas as pd
```

```python
corr_rank = pd.read_csv('MeanPearsonCorr_SameGene.csv')
```

```python
corr_rank
```

```python
corr_rank.columns
```

```python
corr_rank['Difference'] = corr_rank['Overlapping Mean Pearson Correlation'] - corr_rank['Non-overlapping Mean Pearson Correlation']

corr_rank['Normalized_Difference'] = corr_rank.Difference/corr_rank['Difference'].mean()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
fig, ax = plt.subplots(figsize = (7,5))
sns.regplot(data = corr_rank, x = 'Feature Rank', y = 'Normalized_Difference', 
           scatter = False, lowess = True)

ax.set_xscale('log')
```
